### Decoding using LOO-CV

```python
import time

# === Decoding usign LOO-cv===

all_predictions = {}

for area in areas:
    print(f"\n=== Checking sessions for area: {area} ===")
    valid_sessions = []

    for session in sessions:
        area_idx = data['unit'][session]['area'].values == area
        num_neurons = np.sum(area_idx)

        if num_neurons >= 10:
            valid_sessions.append(session)
        else:
            print(f"  ✗ Session {session} skipped → only {num_neurons} neurons")

    if valid_sessions:
        print(f"==> Decoding {len(valid_sessions)} sessions for area {area}")
        start_time = time.time()

        # Using the decoding function to train and test on the same timebin using LOO
        data, area_preds = decode_area_loo_parallel_timebin(
            data, valid_sessions, area, variables_to_analyze
        )
        all_predictions.update(area_preds)

        elapsed = time.time() - start_time
        print(f"✓ Finished decoding area {area} in {elapsed:.2f}s")
    else:
        print(f"No sessions to decode for area {area}.")

# Save predicted trials
with open(os.path.join(results_dir, 'predicted_trial_SVM_all_timebin.pkl'), 'wb') as f:
    pickle.dump(data['trial'], f)

# Save predicted results
with open(os.path.join(results_dir, 'predicted_results_SVM_all_timebin.pkl'), 'wb') as f:
    pickle.dump(all_predictions, f)

print("✓ Saved decoded trials and results to 'Neural/results/'")
```

### Plotting decoded data over time

```python
def plot_decoding_comparison(all_mean_scores, all_sem_scores, time, areas, previous_variables, current_variables):
    """
    Plot decoding performance comparison for neural data across brain areas and task variables.

    Parameters:
    -----------
    all_mean_scores : dict
        Nested dictionary of mean decoding performance. Structure: area -> variable -> array of scores over time.
    all_sem_scores : dict
        Same structure as all_mean_scores but for SEM (standard error of the mean).
    time : array-like
        Time axis values.
    areas : list of str
        List of brain areas to include in the plot.
    previous_variables : list of str
        Names of variables from the previous trial to decode.
    current_variables : list of str
        Names of variables from the current trial to decode.

    Output:
    -------
    Saves a multi-panel figure in the 'figures' directory and displays the plot.
    """

    variable_pairs = list(zip(previous_variables, current_variables))
    row_labels = ["Color", "Direction", "Response", "Context"]
    column_labels = ["Previous", "Current"]

    n_rows = len(variable_pairs)
    n_cols = 2

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 12), sharex=True)
    axes = np.atleast_2d(axes)

    area_colors = {
        'PFC': 'b',
        'FEF': 'r',
        'LIP': 'g',
        'Parietal': 'y',
        'IT': 'c',
        'MT': 'm',
        'V4': 'orange'
    }

    # Track y-limits per column
    col_ymins = [np.inf for _ in range(n_cols)]
    col_ymaxs = [-np.inf for _ in range(n_cols)]

    # Plotting
    for i, (prev_var, curr_var) in enumerate(variable_pairs):
        for j, var in enumerate([prev_var, curr_var]):
            ax = axes[i, j]

            for area in areas:
                mean_scores = all_mean_scores[area]
                sem_scores = all_sem_scores[area]

                if var in mean_scores:
                    mean_curve = mean_scores[var]
                    sem_curve = sem_scores[var]

                    if not np.isnan(mean_curve).all():
                        ax.plot(time, mean_curve, label=area, linewidth=3,
                                color=area_colors.get(area, 'black'), zorder=2)
                        ax.fill_between(time, mean_curve - sem_curve, mean_curve + sem_curve,
                                        alpha=0.3, color=area_colors.get(area, 'black'), zorder=1)

                        col_ymins[j] = min(col_ymins[j], np.nanmin(mean_curve - sem_curve))
                        col_ymaxs[j] = max(col_ymaxs[j], np.nanmax(mean_curve + sem_curve))

            ax.axvline(0, color='k', linestyle='--', linewidth=2)

            ax.axvline(0, color='k', linestyle='--', linewidth=2.5)

            # Thicker and bigger tick marks
            ax.tick_params(axis='both', which='major', labelsize=24, width=4, length=12)
            ax.tick_params(axis='both', which='minor', width=3, length=8)

            # Make x and y tick labels bold and larger
            for tick in ax.get_xticklabels() + ax.get_yticklabels():
                tick.set_fontsize(24)
                tick.set_fontweight('bold')

            # Thicker spines
            for spine in ax.spines.values():
                spine.set_linewidth(3)

            # X-axis range restriction
            ax.set_xlim(-1.5, 2.5)

            # Bigger and bolder x-axis label on bottom row
            if i == n_rows - 1:
                ax.set_xlabel("Time (s)", fontsize=26, fontweight='bold')

    # Set shared y-limits
    for j in range(n_cols):
        if np.isfinite(col_ymins[j]) and np.isfinite(col_ymaxs[j]):
            for i in range(n_rows):
                axes[i, j].set_ylim(col_ymins[j], col_ymaxs[j])

    # Column headers
    for j, label in enumerate(column_labels):
        axes[0, j].set_title(label, fontsize=20, fontweight='bold', pad=20)

    # Row labels
    # Row labels 
    fig.canvas.draw()
    row_y_positions = [np.mean([ax.get_position().y0, ax.get_position().y1]) for ax in axes[:, 0]]
    for y_pos, label in zip(row_y_positions, row_labels):
        fig.text(0.035, y_pos, label, va='center', ha='left', fontsize=18, fontweight='bold', rotation='vertical')

    # Shared y-axis label (move slightly further left)
    fig.text(0.015, 0.5, "Decoding Performance", va='center', ha='center',
             rotation='vertical', fontsize=28, fontweight='bold')


    # Shared legend
    handles, labels = axes[0, 0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='lower center', ncol=len(areas),
               fontsize=15, frameon=False, title='Area', title_fontsize=17)

    # Save figure
    fig_path = os.path.join(figures_dir, "decoding_poster_plot_all_time_bin.svg")
    fig.savefig(fig_path, format='svg', bbox_inches='tight')

    plt.show()
```

```python

# === Compute decoding accuracies ===
all_mean_scores = defaultdict(dict)
all_sem_scores = defaultdict(dict)

for area in all_predictions.keys():
    for session in all_predictions[area].keys():
        preds = all_predictions[area][session]

        for var in preds.keys():
            var_preds = preds[var]  # shape: (n_trials, n_timebins)

            trial_data = trials[session]
            true_labels = trial_data[var].values  # (n_trials,)

            correlations = []
            for t in range(var_preds.shape[1]):  # for each time bin
                pred_t = var_preds[:, t]  # predictions at time t
                mask = ~np.isnan(pred_t) & ~np.isnan(true_labels)

                if np.sum(mask) > 2:
                    r, _ = stats.pearsonr(pred_t[mask], true_labels[mask])
                else:
                    r = np.nan

                correlations.append(r)

            correlations = np.array(correlations)

            if var not in all_mean_scores[area]:
                all_mean_scores[area][var] = []
                all_sem_scores[area][var] = []

            all_mean_scores[area][var].append(correlations)
            all_sem_scores[area][var].append(correlations)

# Average across sessions
for area in all_mean_scores.keys():
    for var in all_mean_scores[area].keys():
        stacked = np.stack(all_mean_scores[area][var], axis=0)
        all_mean_scores[area][var] = np.nanmean(stacked, axis=0)
        all_sem_scores[area][var] = np.nanstd(stacked, axis=0) / np.sqrt(np.sum(~np.isnan(stacked), axis=0))
```

```python
# === Define time correctly ===
time = np.arange(-2.5, 3.5, bin_size)[:239]

# Plotting the data over_time
plot_decoding_comparison(
    all_mean_scores,
    all_sem_scores,
    time,
    areas,
    previous_variables,
    current_variables
)

```
